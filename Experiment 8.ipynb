```python
#Declaration of NumPy
import numpy as np
a = np.array([10, 20, 30, 40])
b = np.array([[1,2,3], [4,5,6]])
print(a)
print(b)
```

```python
#Dimension of Array
print(a.ndim)
print(b.ndim)

```

```python
#Shape (How many members are there)
print(a.shape)
print(b.shape)
```

```python
#Datatype (Integer form)

print(a.dtype)
print(b.dtype)
```

```python
#Matrix Multiplication

print(np.zeros((2,3)))   #2,3 shows the size, 2 rows and 3 columns

```

```python
print(np.zeros((3,3)))    # 3 rows and 3 columns
```

```python
print(np.ones((3,3)))
```

```python
print(np.eye((3)))   #Diagonal Matrix
```

```python
print(np.arange(1,10,2))
```
